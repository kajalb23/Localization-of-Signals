```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# 1. Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Normalize input data
x_train = x_train / 255.0
x_test = x_test / 255.0

# 3. Convert labels to one-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 4. Build DNN model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# 5. Compile model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ✅ 6. Train model and store in history
history = model.fit(x_train, y_train, epochs=5, validation_split=0.2)

# 7. Evaluate model
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")

# ✅ 8. Plot accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')

plt.title('Epoch vs Accuracy (Epoch=5)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import tensorflow as tf
import matplotlib.pyplot as plt  # ✅ Needed for plotting

# Step 1: Load full dataset
(x_full, y_full), _ = tf.keras.datasets.mnist.load_data()

# Step 2: Normalize and take only first 500 samples
x_subset = x_full[:500] / 255.0
y_subset = y_full[:500]

# Step 3: Split into train (400) and test (100)
x_train, x_test = x_subset[:400], x_subset[400:]
y_train, y_test = y_subset[:400], y_subset[400:]

# Step 4: Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10)
])

# Step 5: Compile
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Step 6: Train the model (with 20% of train used as validation)
history = model.fit(x_train, y_train, epochs=5, validation_split=0.2)

# Step 7: Evaluate on held-out test set
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"\n✅ Test Accuracy: {test_acc:.4f}")

# Step 8: Plot accuracy curves
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Epoch vs Accuracy (Epoch=5)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

```
